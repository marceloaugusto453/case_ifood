# Qual √© a distribui√ß√£o de sal√°rio na nossa base? Temos clientes que ganham muito bem e outros que n√£o ganham muito bem? Veja na coluna Income

````python
#vendo o histograma
df["Income"].hist()
````

````python
#vendo o boxplot - voc√™s podem usar a biblioteca PLOTLY tamb√©m
sns.boxplot(df["Income"])
````

Al√©m de olhar o boxplot e o histograma, vamos calcular o coeficiente de varia√ß√£o (CV) da vari√°vel Income. O CV ajuda a entender o qu√£o espalhados os valores est√£o em rela√ß√£o a m√©dia, sendo util para comparar variabilidades entre diferentes conjuntos de dados. Ele geralmente √© expresso como um n√∫mero decimal ou em porcentagem, onde valores mais pr√≥ximos de 1 (ou 100%) indicam uma maior dispers√£o dos dados.

````python
# calcular o coeficiente de varia√ß√£o
df["Income"].std()/df["Income"].mean()
````

Para compreender melhor a distrbui√ß√£o dos dados e verificar se eles se aproximam de uma curva normal, podemos utilizar o calculo do skewness. O skewness fornece um calculo para verificarmos o quanto nossos dados est√£o distribuidos simetricamente em rela√ß√£o √† m√©dia. Este valor geralmente fica entre -1 e 1. Quanto mais pr√≥ximos das extremidades deste intervalo, mais os dados s√£o assim√©tricos. Podemos resumir da seguinte maneira:
- skew > 0 : assimetria positiva (√† direita). A calda da curva ser√° mais longa para valores maiores do eixo x.
- skew < 0: assimetria negativa (√† esquerda). A calda da curva ser√° mais longa para valores menores do eixo x.
- skew = 0: Distribui√ß√£o sim√©trica (Curva normal).

````python
df["Income"].skew()
````

Os salarios parecem bem distribuidos na nossa base. Algumas poucas pessoas ganham pouco e algumas poucas ganham muito, por√©m n√£o h√° outliers.
A dispers√£o por outro lado √© grande (muitos dados se afastam da m√©dia).
Pelo skew e pelo boxplot, podemos dizer que apesar de n√£o completamente normal, a curva se aproxima de uma normal.

# As pessoas gastam mais ou menos em nossa plataforma quando t√™m filhos?


````python
sns.boxplot(x='kids', y='expenses', data=df)
````

````python
#calculando a m√©dia de gastos
df.groupby("kids")["expenses"].mean()
````

````python
#calculando a mediana
df.groupby("kids")["expenses"].median()
````

Podemos ver que tanto a m√©dia quanto a mediana 0 e 1 filho √© maior nessa amostra de dados, por√©m a m√©dia de 2 filhos √© menor do que de 3 filhos, enquanto a mediana de 2 filhos √© maior do que de 3 filhos.
Ambos os grupos 2 e 3 tem m√©dias bem maiores do que a mediana, indicando outliers superiores (o pr√≥prio gr√°fico j√° mostra isso). Grupo 1 tamb√©m tem outliers superiores, mas essa diferen√ßa √© menos acentuada.
Pessoas com nenhum filho tem uma distribui√ß√£o um pouco mais similar a uma normal (sem outliers, e mediana levemente centrada do meio do boxplot)
